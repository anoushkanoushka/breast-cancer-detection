```python
import os
import numpy as np
import pandas as pd
from PIL import Image
from pathlib import Path
```

```python
# Find your dataset folder in Windows Explorer
# Click the address bar at the top and copy the path
# Paste it below between the quotes

DATASET_PATH = Path(r"C:\Users\jessi\OneDrive\Desktop\aiims2")

# Example:
# DATASET_PATH = Path(r"C:\Users\John\Downloads\Dataset_BUSI_with_GT")

LABELS = ['benign', 'malignant', 'normal']

# Quick check — prints what it finds
for label in LABELS:
    folder = DATASET_PATH / label
    files = list(folder.glob("*.png"))
    images = [f for f in files if '_mask' not in f.name]
    print(f"{label}: {len(images)} images found")
```

```python
records = []

for label in LABELS:
    folder = DATASET_PATH / label
    
    # Get all images (exclude masks)
    images = sorted([
        f for f in folder.glob("*.png") 
        if '_mask' not in f.name
    ])
    
    for img_path in images:
        # Find corresponding mask
        mask_name = img_path.stem + '_mask.png'
        mask_path = folder / mask_name
        
        if not mask_path.exists():
            print(f"WARNING: No mask found for {img_path.name}")
            continue
        
        # Load image and mask
        img  = np.array(Image.open(img_path).convert('L'))   # grayscale
        mask = np.array(Image.open(mask_path).convert('L'))  # grayscale
        
        # Binary mask (some masks are 0-255, convert to 0-1)
        mask_binary = (mask > 127).astype(np.uint8)
        
        # Auto-extract basic metadata
        h, w = img.shape
        total_pixels     = h * w
        lesion_pixels    = mask_binary.sum()
        lesion_area_pct  = round(lesion_pixels / total_pixels * 100, 2)
        
        # Image intensity stats
        mean_intensity   = round(img.mean(), 2)
        std_intensity    = round(img.std(), 2)
        min_intensity    = int(img.min())
        max_intensity    = int(img.max())
        
        # Lesion region stats
        if lesion_pixels > 0:
            lesion_mean = round(img[mask_binary == 1].mean(), 2)
            lesion_std  = round(img[mask_binary == 1].std(), 2)
        else:
            lesion_mean = None
            lesion_std  = None
        
        records.append({
            'case_id'          : img_path.stem,
            'label'            : label,
            'image_path'       : str(img_path),
            'mask_path'        : str(mask_path),
            'image_width'      : w,
            'image_height'     : h,
            'mean_intensity'   : mean_intensity,
            'std_intensity'    : std_intensity,
            'min_intensity'    : min_intensity,
            'max_intensity'    : max_intensity,
            'lesion_pixels'    : int(lesion_pixels),
            'lesion_area_pct'  : lesion_area_pct,
            'lesion_mean'      : lesion_mean,
            'lesion_std'       : lesion_std,
            # --- fill these in manually later ---
            'pixel_size_mm'    : None,
            'BIRADS'           : None,
            'echogenicity'     : None,   # hypo / iso / hyper / anechoic
            'posterior_feature': None,   # shadowing / enhancement / none
            'shape'            : None,   # oval / round / irregular
            'margin'           : None,   # circumscribed / indistinct / spiculated
        })

df = pd.DataFrame(records)
print(f"Total cases: {len(df)}")
df.head()
```

```python
# Shows any cases where mask was missing
missing = df[df['lesion_pixels'] == 0]
print(f"Cases with empty/missing mask: {len(missing)}")
missing[['case_id', 'label']]
```

```python
print("=== Cases per label ===")
print(df['label'].value_counts())

print("\n=== Image size distribution ===")
print(df[['image_width', 'image_height']].describe())

print("\n=== Lesion area % by label ===")
print(df.groupby('label')['lesion_area_pct'].describe().round(2))
```

```python
OUTPUT_PATH = DATASET_PATH / 'metadata.csv'
df.to_csv(OUTPUT_PATH, index=False)
print(f"Saved to: {OUTPUT_PATH}")
```
